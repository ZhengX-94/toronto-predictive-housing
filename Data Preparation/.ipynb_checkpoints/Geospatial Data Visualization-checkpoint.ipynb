```python
import pandas as pd

import plotly.graph_objs as go

import plotly.offline as py

py.init_notebook_mode(connected=True)
```

```python
houses = pd.read_excel('data.xlsx', index_col=0)

# Adding province name and country to be precise
houses[['full_address']] = houses[['full_address']].astype(str) + ', Ontario, Canada'
houses.head(5)
```

```python
df = houses.copy()
df['final_price'] = df['final_price'].apply(lambda x: x.replace(',',''))
df['bedrooms'] = df['bedrooms'].apply(lambda x: x[0])
df['bathrooms'] = df['bathrooms'].apply(lambda x: x[0])
df['parking'] = df['parking'].apply(lambda x: x[0] if x[0] !='n' else 0)

```

```python
house_types = df['type'].value_counts()


data = go.Bar(x=house_types.index, y=house_types.values)

layout = go.Layout(
    title='House types',
)

fig = go.Figure(data=[data], layout=layout)

py.iplot(fig)
```

```python
# Avg price

house_price = df.groupby('type')['final_price'].mean().sort_values(ascending=False)/1000
house_price

data = go.Bar(x=house_price.index, y=house_price.values)

layout = go.Layout(
    title='House average final price(k$)',
)

fig = go.Figure(data=[data], layout=layout)

py.iplot(fig)
```
